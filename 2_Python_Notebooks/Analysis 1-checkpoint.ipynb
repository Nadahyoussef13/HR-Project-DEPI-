```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.formula.api as smf

hr = pd.read_csv('C:/Users/LOQ/Desktop/HR Project/HRVisual.csv')
hr['YearsAtCompany'] = pd.to_numeric(hr['YearsAtCompany'], errors='coerce')
hr['Salary_Cleaned'] = pd.to_numeric(hr['Salary_Cleaned'], errors='coerce')

corr = hr['YearsAtCompany'].corr(hr['Salary_Cleaned'])
print(f"Pearson correlation (YearsAtCompany vs Salary_Cleaned): {corr:.3f}")

hr['TenureGroup'] = pd.cut(hr['YearsAtCompany'], bins=[-0.1,2,5,10],
                           labels=['<2 yrs','2-5 yrs','5-10 yrs'])
agg = hr.groupby(['TenureGroup','EducationLevel.txt'])['Salary_Cleaned'].agg(['mean','count']).reset_index()
agg['mean'] = agg['mean'].round(0)
display(agg.sort_values(['TenureGroup','mean'], ascending=[True,False]))

groups = [group['Salary_Cleaned'].dropna() for name, group in hr.groupby('TenureGroup')]
anova_res = stats.f_oneway(*groups)
print(f"F = {anova_res.statistic:.3f}, p = {anova_res.pvalue:.4f}")


hr['EducationLevel_txt_clean'] = hr['EducationLevel.txt'].astype(str)
model = smf.ols('Salary_Cleaned ~ YearsAtCompany + C(EducationLevel_txt_clean)', data=hr).fit()
print(model.summary())


plt.figure(figsize=(8,5))
sns.scatterplot(data=hr.sample(frac=0.5, random_state=1), x='YearsAtCompany', y='Salary_Cleaned', alpha=0.5)
sns.regplot(data=hr, x='YearsAtCompany', y='Salary_Cleaned', scatter=False, lowess=True, color='red')
plt.title('Salary vs Years at Company (LOWESS trend)')
plt.xlabel('Years at Company')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()
```
