```python
perf_env = pd.merge(
    performance,
    satisfaction,
    left_on="EnvironmentSatisfaction",
    right_on="SatisfactionID",
    how="left"
).rename(columns={"SatisfactionLevel": "EnviromentSatisfactionLevel"})
```

```python
perf_job = pd.merge(
    performance,
    satisfaction,
    left_on="JobSatisfaction",
    right_on="SatisfactionID",
    how="left"
).rename(columns={"SatisfactionLevel": "JobSatisfactionLevel"})
```

```python
perf_Relation = pd.merge(
    performance,
    satisfaction,
    left_on="RelationshipSatisfaction",
    right_on="SatisfactionID",
    how="left"
).rename(columns={"SatisfactionLevel": "RelationshipSatisfactionLevel"})
```

```python
perf_worklife = pd.merge(
    performance,
    satisfaction,
    left_on="WorkLifeBalance",
    right_on="SatisfactionID",
    how="left"
).rename(columns={"SatisfactionLevel": "worklifebalancesatisfaction"})
```

```python
perf_with_self = pd.merge(
    perf_worklife,
    ratings,
    left_on="SelfRating",
    right_on="RatingID",
    how="left"
).rename(columns={"RatingLevel": "SelfRatingLevel"})
```

```python
perf_all = pd.merge(
    perf_with_self,
    ratings,
    left_on="ManagerRating",
    right_on="RatingID",
    how="left"
).rename(columns={"RatingLevel": "SelfRatingLevel"})
```
